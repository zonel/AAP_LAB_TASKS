## 1. Wstęp: Dlaczego testujemy?

Wyobraź sobie, że budujesz dom. Czy wprowadziłbyś się do niego bez inspekcji budowlanej? Pewnie nie — chciałbyś mieć pewność, że ściany są proste, instalacja elektryczna bezpieczna, a dach nie przecieka.

**Testy w programowaniu pełnią dokładnie tę samą rolę** — sprawdzają, czy Twój kod działa zgodnie z oczekiwaniami, zanim trafi na produkcję.

**Trzy kluczowe korzyści z testowania:**
1. **Wczesne wykrywanie błędów** — im wcześniej znajdziesz błąd, tym taniej i szybciej go naprawisz.
2. **Dokumentowanie zachowania** — test jest żywą dokumentacją, która mówi: *"ta funkcja powinna zwracać X dla danych Y"*.
3. **Bezpieczne zmiany** — refaktoryzujesz kod? Testy powiedzą Ci, czy niczego nie zepsułeś.

W Pythonie mamy dwa główne frameworki testowe:
- **unittest** — wbudowany w Pythona (nie trzeba nic instalować)
- **pytest** — zewnętrzna biblioteka (prostszy, bardziej nowoczesny)

Na dzisiejszych zajęciach poznamy oba i porównamy je na praktycznych przykładach.

## 2. Słowo kluczowe `assert`

Zanim przejdziemy do frameworków testowych, poznajmy podstawę: słowo kluczowe **`assert`**.

`assert` sprawdza, czy podany warunek jest prawdziwy. Jeśli nie — rzuca wyjątek `AssertionError`.

```python
assert warunek, "Opcjonalna wiadomość o błędzie"
```

To fundament, na którym opiera się całe testowanie w Pythonie.

````python
# Podstawowe uzycie assert
x = 5
assert x == 5, "x powinno wynosic 5"
print("Asercja przeszla: x == 5")

# Co sie stanie, gdy warunek NIE jest spelniony?
try:
    assert x == 10, "x nie wynosi 10!"
except AssertionError as e:
    print(f"Blad asercji zlapany: {e}")

# Praktyczny przyklad -- testowanie prostej funkcji
def dodaj(a, b):
    return a + b

assert dodaj(2, 3) == 5
assert dodaj(-1, 1) == 0
assert dodaj(0, 0) == 0
print("Wszystkie asercje przeszly pomyslnie!")
````

## 3. unittest — wbudowany framework testowy

**unittest** to moduł wbudowany w Pythona (nie trzeba nic instalować). Jest wzorowany na frameworku JUnit z Javy.

**Kluczowe zasady unittest:**
- Testy organizujemy w **klasy** dziedziczące po `unittest.TestCase`
- Każda metoda testowa **musi zaczynać się od `test_`** (inaczej nie zostanie wykryta!)
- Używamy specjalnych metod asercji zamiast zwykłego `assert`

### Najpopularniejsze asercje unittest:

| Metoda | Sprawdza |
|--------|----------|
| `assertEqual(a, b)` | a == b |
| `assertNotEqual(a, b)` | a != b |
| `assertTrue(x)` | bool(x) is True |
| `assertFalse(x)` | bool(x) is False |
| `assertRaises(exc)` | czy rzucono wyjątek |
| `assertIn(a, b)` | a in b |

### Demo: Testowanie klasy Calculator z unittest

Poniżej znajdziesz klasę `Calculator` oraz jej testy jednostkowe.

**WAŻNE:** Testy unittest najlepiej uruchamiać z terminala (nie z Jupytera), ponieważ `unittest.main()` kończy proces. Dlatego kod klasy i testów znajduje się w osobnych plikach:
- `calculator.py` — klasa, którą testujemy
- `test_calculator_unittest.py` — testy

Poniżej pokazujemy zawartość obu plików, a następnie uruchamiamy testy komendą terminalową.

````python
# ========================================
# Zawartosc pliku: calculator.py
# ========================================

class Calculator:
    """Prosty kalkulator z czterema podstawowymi operacjami."""

    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b

    def multiply(self, a, b):
        return a * b

    def divide(self, a, b):
        if b == 0:
            raise ValueError("Dzielenie przez zero!")
        return a / b

# Szybki test reczny -- upewnijmy sie, ze klasa dziala
calc = Calculator()
print(f"2 + 3 = {calc.add(2, 3)}")
print(f"10 - 5 = {calc.subtract(10, 5)}")
print(f"3 * 7 = {calc.multiply(3, 7)}")
print(f"10 / 2 = {calc.divide(10, 2)}")
print("\nKlasa Calculator dziala poprawnie!")
````

````python
# ========================================
# Zawartosc pliku: test_calculator_unittest.py
# ========================================
# (ponizej pokazujemy kod -- NIE uruchamiamy go tutaj,
#  bo unittest.main() zakonczy proces Jupytera)

print("""
import unittest
from calculator import Calculator


class TestCalculator(unittest.TestCase):

    def setUp(self):
        \"\"\"Tworzy nowa instancje Calculator przed KAZDYM testem.\"\"\"
        self.calc = Calculator()

    def test_add_positive_numbers(self):
        result = self.calc.add(2, 3)
        self.assertEqual(result, 5)

    def test_add_negative_numbers(self):
        result = self.calc.add(-1, -1)
        self.assertEqual(result, -2)

    def test_subtract(self):
        result = self.calc.subtract(10, 5)
        self.assertEqual(result, 5)

    def test_multiply(self):
        result = self.calc.multiply(3, 7)
        self.assertEqual(result, 21)

    def test_divide(self):
        result = self.calc.divide(10, 2)
        self.assertEqual(result, 5.0)

    def test_divide_by_zero_raises_error(self):
        with self.assertRaises(ValueError):
            self.calc.divide(10, 0)


if __name__ == "__main__":
    unittest.main()
""")
````

````python
# Uruchamiamy testy unittest z poziomu terminala
# Flaga -v (verbose) pokazuje szczegolowy wynik kazdego testu
!python -m unittest test_calculator_unittest -v
````

### setUp i tearDown

Zwróć uwagę na metodę `setUp` w powyższym kodzie:

- **`setUp`** — wywoływana **przed każdym testem**. Przygotowuje środowisko (tworzy obiekty, otwiera połączenia, ustawia dane).
- **`tearDown`** — wywoływana **po każdym teście**. Sprząta (zamyka pliki, usuwa dane tymczasowe, przywraca stan).

Dzięki `setUp` każdy test zaczyna od **czystego stanu** — testy są od siebie **niezależne**.

```
setUp()  -->  test_add()       -->  tearDown()
setUp()  -->  test_subtract()  -->  tearDown()
setUp()  -->  test_multiply()  -->  tearDown()
...
```

Każdy test dostaje **świeży** obiekt `Calculator()` — nawet jeśli jeden test go zmodyfikuje, następny nie odczuje tego efektu.

## 4. pytest — nowoczesny framework testowy

**pytest** to zewnętrzna biblioteka, która upraszcza pisanie testów. Kluczowe różnice vs unittest:

| Cecha | unittest | pytest |
|-------|----------|--------|
| Struktura | Klasy + metody | Zwykłe funkcje |
| Asercje | `self.assertEqual(a, b)` | `assert a == b` |
| Przygotowanie | `setUp` / `tearDown` | `@pytest.fixture` |
| Instalacja | Wbudowany | `pip install pytest` |
| Komunikaty błędów | Ogólne | Szczegółowe (co dokładnie się różni) |

**Główna zaleta pytest:** mniej kodu, prostszy zapis, lepsze komunikaty błędów.

Instalacja: `pip install pytest`

### Demo: Te same testy Calculator w pytest

Porównaj poniższy kod z wersją unittest — ten sam kalkulator, ale testy są krótsze i bardziej czytelne.

Plik: `test_calculator_pytest.py`

````python
# ========================================
# Zawartosc pliku: test_calculator_pytest.py
# ========================================

print("""
import pytest
from calculator import Calculator


@pytest.fixture
def calc():
    \"\"\"Fixture zastepujacy setUp -- tworzy Calculator dla kazdego testu.\"\"\"
    return Calculator()


def test_add(calc):
    assert calc.add(2, 3) == 5
    assert calc.add(-1, 1) == 0


def test_subtract(calc):
    assert calc.subtract(10, 5) == 5


def test_multiply(calc):
    assert calc.multiply(3, 7) == 21


def test_divide(calc):
    assert calc.divide(10, 2) == 5.0


def test_divide_by_zero(calc):
    with pytest.raises(ValueError):
        calc.divide(10, 0)
""")
````

````python
# Uruchamiamy testy pytest -- zwroc uwage na kolorowy, czytelny output
!pytest test_calculator_pytest.py -v
````

### Fixtures — przygotowanie środowiska testowego

**Fixture** w pytest to funkcja oznaczona dekoratorem `@pytest.fixture`. Zastępuje `setUp` z unittest, ale jest bardziej elastyczna:

- Może być **współdzielona** między wieloma plikami testowymi (przez plik `conftest.py`)
- Może zwracać dane lub obiekty
- Fixture może używać innych fixtures (kompozycja)

**Jak to działa?** Pytest widzi, że funkcja testowa przyjmuje argument `calc` — szuka fixture o tej nazwie i automatycznie go przekazuje:

```python
@pytest.fixture
def calc():                    # <-- definiujemy fixture
    return Calculator()

def test_add(calc):            # <-- pytest automatycznie przekazuje fixture
    assert calc.add(2, 3) == 5
```

````python
# Demonstracja koncepcji fixture -- prosty przyklad w Jupyterze
# (w prawdziwych testach uzywamy dekoratora @pytest.fixture)

def przygotuj_liste_zakupow():
    """Symulacja fixture -- przygotowuje dane testowe."""
    return ["mleko", "chleb", "maslo", "jajka"]

# "Testy" uzywajace przygotowanych danych
dane = przygotuj_liste_zakupow()

assert len(dane) == 4, "Lista powinna miec 4 elementy"
assert "mleko" in dane, "Mleko powinno byc na liscie"
assert "woda" not in dane, "Wody nie powinno byc na liscie"

print(f"Lista zakupow: {dane}")
print(f"Liczba elementow: {len(dane)}")
print("Wszystkie sprawdzenia przeszly!")
````

### Parametryzacja — jeden test, wiele zestawów danych

**`@pytest.mark.parametrize`** pozwala uruchomić ten sam test z różnymi danymi wejściowymi. To eliminuje konieczność pisania wielu prawie identycznych testów.

Dekorator przyjmuje dwa argumenty:
1. Nazwy parametrów jako string (oddzielone przecinkami)
2. Listę krotek z wartościami

```python
@pytest.mark.parametrize("a, b, expected", [
    (2, 3, 5),       # przypadek 1
    (-1, 1, 0),      # przypadek 2
    (0, 0, 0),       # przypadek 3
])
def test_add(a, b, expected):
    calc = Calculator()
    assert calc.add(a, b) == expected
```

Jeden test, ale pytest uruchomi go **3 razy** — osobno dla każdego zestawu danych!

````python
# Uruchamiamy TYLKO sparametryzowany test -- zwroc uwage na output
# Kazdy zestaw danych jest pokazany osobno!
!pytest test_calculator_pytest.py::test_add_parametrized -v
````

### Marks — oznaczanie testów

**Marks** (znaczniki) pozwalają kontrolować, które testy uruchomić i jak się mają zachować:

- **`@pytest.mark.skip(reason="...")`** — pomija test bezwarunkowo. Przydatne gdy funkcjonalność jeszcze nie jest zaimplementowana.
- **`@pytest.mark.xfail(reason="...")`** — oczekuje niepowodzenia. Test może nie przejść, a całość i tak będzie "zielona". Przydatne dla znanych bugów.

```python
@pytest.mark.skip(reason="Funkcja jeszcze nie zaimplementowana")
def test_future_feature():
    assert new_function() == expected

@pytest.mark.xfail(reason="Znany bug #123")
def test_known_bug():
    assert broken_function() == expected
```

````python
# Uruchamiamy WSZYSTKIE testy pytest -- zwroc uwage na:
# - 's' = skipped (pominiete)
# - 'x' = xfail (oczekiwane niepowodzenie)
# - 'PASSED' = test przeszedl
!pytest test_calculator_pytest.py -v
````

### Struktura projektu testowego

Dobre praktyki organizacji testów — testy powinny odwzorowywać strukturę projektu:

```
moj_projekt/
├── moj_modul/
│   ├── __init__.py
│   ├── calculator.py
│   └── product.py
├── tests/
│   ├── __init__.py
│   ├── test_calculator.py
│   └── test_product.py
├── conftest.py          <-- fixtures wspólne dla wszystkich testów
├── pytest.ini           <-- konfiguracja pytest
└── requirements.txt
```

**Zasady:**
- Pliki testowe zaczynają się od `test_` (np. `test_calculator.py`)
- Każdy moduł ma odpowiadający mu plik testowy
- `conftest.py` zawiera fixtures dostępne we wszystkich testach (nie trzeba go importować!)
- Uruchomienie wszystkich testów: `pytest` (z katalogu głównego projektu)
